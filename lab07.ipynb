## Zadanie 2

Zadanie należy wykonać wykorzystując program `BagOfWords` z poprzednich zajęć.


Przerób ją tak, żeby interpunkcja, cyfry i wszelkie inne znaki nie przeszkadzały w parsowaniu tekstu. Uruchom ja na [tekście hamleta](http://www.gutenberg.org/cache/epub/1787/pg1787.txt). Ile razy występuje słowo hamlet? Jak brzmi 10 najczęściej występujących słów?

```python
from collections import defaultdict
import re
import json

class BagOfWords(object):

    def __init__(self, content=None, dict=None):
        if content is None and dict is None:
            raise AttributeError

        if dict:
            self.__words = dict
            return

        if not isinstance(content, str):
            content = content.read()

        content = re.sub('[^a-z\s]', '', content.lower())
        content = re.sub('\s+', ' ', content)

        words = defaultdict(int)
        for word in content.split(' '):
            words[word] += 1

        self.__words = words

    def __iter__(self):
        self.__iterator = iter(
            sorted(self.__words.items(), key=lambda item: item[1], reverse=True))
        return self

    def __next__(self):
        return next(self.__iterator)[0]

    def __repr__(self):
        return ", ".join((f'{k}:{v}' for k,v in self.__words.items()))

    def __contains__(self, item):
        return item in self.__words

    def __add__(self, other):
        a = self.__words
        b = other.__words
        return BagOfWords(dict={key: a[key] + b[key] if key in a and key in b else value for d in [a, b] for key, value in d.items()})

    def __getitem__(self, key):
        return self.__words[key]

    def __setitem__(self, key, value):
        self.__words[key] = value

    def top_ten(self):
        return ', '.join(f'{k}:{v}' for k,v in sorted(self.__words.items(), key=lambda x: x[1], reverse=True)[:10])

    def save(self, filename):
        with open(filename, "w") as outfile:
            json.dump(self.__words, outfile)

    def load(self, filename):
        with open(filename, "r") as infile:
            self.__words = json.load(infile)


hamlet = open('hamlet.txt')
h = BagOfWords(hamlet)

print(h.top_ten())

h.save('bagofwordss.json')
h.load('example.json')

print(h.top_ten())
```

- przeprowadź symulacyjne wyświetlanie kolejnych krajów w miarę jak w danym kraju wybija północ, opóżnienie wyświetlania ustaw proporcjonalne do realnego czasu

```python
from countries import countries
from datetime import datetime
import pytz

sortedList = sorted(countries, key=lambda x: x['continent'])
for country in sortedList:
    for tzone in country['timezones']:
        tz = pytz.timezone(tzone)
        time = datetime.now(tz)
        print(f"{tzone} - {country['name']} - {country['continent']} - {time}")

while True:
    for country in sortedList:
        for tzone in country['timezones']:
            tz = pytz.timezone(tzone)
            time = datetime.now(tz)
            if time.hour == 24 and time.minute == 00:
                print(f"{tzone} - {country['name']} - {country['continent']} - {time}")

```
